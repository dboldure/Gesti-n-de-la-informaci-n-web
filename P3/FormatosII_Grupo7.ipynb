Para facilitar la selección del parque se puede asignar a cada parque un número. Después de haber elegido un parque y habérsele mostrado la información, se le volverá a mostrar el menú de jardines por si quiere consultar otro parque. Entre las opciones del menú deberá existir una que sea para finaliza la búsqueda.

Observar que no en todos los parques del fichero aparecen todos los campos de información requeridos. Por ejemplo,hay parques que no tienen horario o no tienen equipamiento.

```python
import urllib.request
from xml.etree import ElementTree as ET

def obtener_Parques():
    nFichero = "parques.xml"
    x = urllib.request.urlopen('https://datos.madrid.es/egob/catalogo/200761-0-parques-jardines.xml')
    a=x.read()
    a=a.decode("utf-8")
    prueba=open(nFichero,"w")
    prueba.write(str(a))
    prueba.close()
    
    
def mostrarParques():
    dicc_parques ={}
    archivo =open("parques.xml","rt")
    arbol=ET.parse(archivo)
    print("Opcion 0: Salida \n")
    print("Parques disponibles:")
        
    i=1
    for nodo in arbol.iter("contenido"):
        for dato in nodo:
            if dato.tag == "atributos":
                for valor in dato:
                    if valor.get("nombre") == "NOMBRE":
                        print("Parque ",i, ": ", valor.text)
                dicc_parques[i]=dato
        i=i+1
    archivo.close()

    return dicc_parques
    
def pedir_parque_y_seleccionar(dicc_parq):
    valor_parque = input("Escoja uno de los parques: \n")
    while not valor_parque.isdigit() or int(valor_parque) < 0 or int(valor_parque) > len(dicc_parq):
        print("El numero escogido no existe, elige uno correcto o pulse 0 para finalizar.")
        valor_parque = input("Escoja uno de los parques: \n")   
    return valor_parque
    
def mostrarSalida(parque):
    for parq in parque:
        if not parq.text == None:
            print ("\n", parq.get("nombre"),":\n",parq.text)
            for atributo in parq:
                print (atributo.get("nombre"),": ",atributo.text)
                
    print("\n")
            
def menuPrincipal():
    fin = False
    obtener_Parques()
    while fin == False:
        dicc_parq = mostrarParques()
        parque=pedir_parque_y_seleccionar(dicc_parq)
        if int(parque)==0: 
            fin = True
            print("Gracias.")
        else:
            mostrarSalida(dicc_parq[int(parque)])
```

```python
menuPrincipal()
```

Este ejercicio se pide crear un buscador sobre la información recuperada, de forma que el usuario podrá buscar parques por diferentes criterios:

 * Accesibilidad
 
 * Nombre
 
 * Barrio
 
 * Distrito
 
Como resultado debería mostrarse un listado de todos los parques que cumplen las condiciones impuestas por el usuario. Se mostrará de cada parque la misma información que en el caso anterior.  En caso de no existir un parque con las condiciones dada, se mostrará un mensaje informativo.

Observar que habrá que preguntar al usuario por cada uno de los criterios, y éste tendrá que seleccionar un valor por cada criterio o no seleccionarlo. Después de lista la información habérsele mostrado la información, se le volverá a preguntar si quiere seguir buscando. Entre las opciones del menú deberá existir una que sea para finaliza la búsqueda.

```python
from xml.etree import ElementTree
import urllib.request

def procesarFichero():
    #Cargamos el contenido de la url
    x = urllib.request.urlopen('https://datos.madrid.es/egob/catalogo/200761-0-parques-jardines.xml')
    a=x.read()
    a=a.decode("utf-8")
    prueba=open("temp.xml","w")
    
    #Guardamos todo el contenido en un fichero temporal, con el que trabajaremos
    prueba.write(str(a))
    prueba.close()

    #Cargamos el fichero temporal para su procesamiento con ElementTree
    fichero=open("temp.xml","rt")
    arbol=ElementTree.parse(fichero)

    #Diccionario para guardar todos los parques y sus datos
    diccionarioParques = {}

    #Para cada parque
    for parque in arbol.iter("contenido"):
        nombre = ""
        for nodo in parque:
            for atributo in nodo:
                #Para cada atributo del parque
                if atributo.tag == "atributo" and atributo.get("nombre") != "ID-ENTIDAD":
                    #Si es el nombre creamos un nuevo parque con key = nombre
                    if atributo.get("nombre") == "NOMBRE":
                        nombre = str(atributo.text)
                        diccionarioParques[nombre]={}
                    #Si no es el nombre metemos cada atributo en el dicionario como key=tipoDato value=contenido ej: {ACCESIBILIDAD:2}, 
                    else:
                        #Si el atributo es un tipo compuesto por mas datos, el valor será un diccionario con dichos datos, ej: {LOCALIZACION:{NOMBRE-VIA:PRINCIPE DE VERGARA},{CLASE-VIAL:CALLE}...}
                        if len(atributo) > 0:
                            key = atributo.get("nombre")
                            diccionarioParques[nombre][key]={}
                            for subAtributo in atributo:
                                diccionarioParques[nombre][key][subAtributo.get("nombre")]=str(subAtributo.text)
                        #Si es un tipo simple, se introduce de forma normal, ej: {ACCESIBILIDAD:2}
                        else:
                            key = atributo.get("nombre")
                            value = str(atributo.text)
                            diccionarioParques[nombre][key]=value

    fichero.close()

    return diccionarioParques

#Menu para que el usuario introduzca los parametros de busqueda
def menuCriterios():
    print("Buscar parques según los siguientes criterios.")
    print("Si desea bucar por alguno de los siguientes criterios complete el campo especificado,")
    print("si por el contrario no desea buscar por alguno de ellos, no lo complete y pulse enter.")
    
    valorAccesibilidad = input("Accesibilidad:")
    valorNombre = input("Nombre:")
    valorBarrio = input("Barrio:")
    valorDistrito = input("Distrito:")

    return [valorAccesibilidad, valorNombre, valorBarrio, valorDistrito]


def buscarParques(listaParametros, diccionarioParques):
    import copy
    existenParametros=False

    listaResultados=[]

    #Si se ha introducido un nombre
    if listaParametros[1]:
        existenParametros=True
        try:
            listaResultados.append({listaParametros[1]:diccionarioParques[listaParametros[1]]})
        except:
            print("No existe ningun parque con el nombre especificado")
    #Si no se ha introducido un nombre, se introducen todos los elementos en la lista
    else:
        for nombre, datos in diccionarioParques.items():
            listaResultados.append({nombre:datos})

    #print(listaResultados)
    lista_aux = copy.deepcopy(listaResultados)
    for parque in listaResultados:
        for nombre, datos in parque.items():
            estaEliminado = False
            #Si se ha introducido accesibilidad
            if listaParametros[0] and not estaEliminado:
                existenParametros=True
                #Si tiene elemento accesibilidad, sino se borra
                if "ACCESIBILIDAD" in datos:
                    if datos["ACCESIBILIDAD"] != listaParametros[0]:
                        lista_aux.remove(parque)
                        estaEliminado = True
                else:
                    lista_aux.remove(parque)
                    estaEliminado = True
                    
    
            #Si se ha introducido Barrio
            if listaParametros[2] and not estaEliminado:
                existenParametros=True
                #Si tiene elemento barrio, sino se borra
                if "BARRIO" in datos["LOCALIZACION"]:
                    if datos["LOCALIZACION"]["BARRIO"] != listaParametros[2]:
                        lista_aux.remove(parque)
                        estaEliminado = True
                else:
                    lista_aux.remove(parque)
                    estaEliminado = True
            
            #Si se ha introducido Distrito
            if listaParametros[3] and not estaEliminado:
                existenParametros=True
                #Si tiene elemento distrito, sino se borra
                if "DISTRITO" in datos["LOCALIZACION"]:
                    if datos["LOCALIZACION"]["DISTRITO"] != listaParametros[3]:
                        lista_aux.remove(parque)
                        estaEliminado = True
                else:
                    lista_aux.remove(parque)
                    estaEliminado = True

    if not existenParametros:
        print("No se han expecificado parametros")
    else:
        if len(lista_aux) > 0:
            for elem in lista_aux:
                mostrarParque(elem)
        else:
            print("No existe ningun parque coincidente con los parametros introducidos")
        
def mostrarParque(parque):
    for nombre, datos in parque.items():
        print("NOMBRE: " + nombre + ": \n")
        for nombre_dato, info in datos.items():
            if info != "None":             
                if type(info) is dict:
                    print("\t" + nombre_dato + ": ")
                    for nombre_datos, info2 in info.items():
                        print("\t\t" + nombre_datos + ": " + info2)
                else:
                    print("\t" + nombre_dato + ": " + info)
    print("\n")
                    
def app():
    diccionarioParques=procesarFichero()
    continuar = True
    while continuar:
        criterios = menuCriterios()
        buscarParques(criterios,  diccionarioParques)
        correcto = False
        while correcto == False:
            respuesta = input("¿Quiere buscar otro parque? S/N")
            if respuesta == "S":
                continuar = True
                correcto = True
            elif respuesta == "N":
                continuar = False
                correcto = True
    print("Gracias")
          
        
```
